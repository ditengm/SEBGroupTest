```python
lst = [26, 13, 6, 16, 17, 2, 0, 1]
```

```python
def spliting(lst):
    ln = len(lst)
    if (ln % 2) == 0:
        return sum(lst[:int(ln/2)]) > sum(lst[int(ln/2):])
    else:
        return sum(lst[:int(ln/2)]) > sum(lst[int(ln/2)+1:])
```

```python
spliting(lst)
```

```python
lst = [
    ['Alexey', 23],
    ['Inna', 26],
    ['Jane', 37],
    ['Olga', 42]
      ]
```

```python
def output_info(lst: list) -> list:
    while True:
        try:
            print("1 -     ")
            print("2 - ")
            print("     ,    1")
            print("    ,    2")
    
            n = int(input()) 
            if (n == 1) or (n == 2):
                break
            else:
                print("     !")
                print()
                continue
        except ValueError:
            print("   !")
            print()
            
    if n == 1:
        return sorted([info[0] for info in lst])
    elif n == 2:
        return sorted([info[1] for info in lst])
        
```

```python
output_info(lst)
```

```python
output_info(lst)
```
